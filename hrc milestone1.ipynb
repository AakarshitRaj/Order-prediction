```python
df = df.fillna(np.NaN)
```

```python
print(df)
```

```python
df["ORDER_CREATION_DATE"] = pd.to_datetime(df["ORDER_CREATION_DATE"], format="%Y%m%d")
```

```python
df["REQUESTED_DELIVERY_DATE"] = pd.to_datetime(df["REQUESTED_DELIVERY_DATE"], format="%Y%m%d")
```

```python
num_higher_date = (df["ORDER_CREATION_DATE"] > df["REQUESTED_DELIVERY_DATE"]).sum()
print("Number of records with order date greater than delivery date:", num_higher_date)
```

```python
df = df[df["ORDER_CREATION_DATE"] <= df["REQUESTED_DELIVERY_DATE"]]
df
```

```python
num1 = df["ORDER_AMOUNT"].str.contains("-", regex=False).sum()
print("Number of records with '-' in the 'ORDER_AMOUNT' field:", num1)
```

```python
df["ORDER_AMOUNT"] = df["ORDER_AMOUNT"].replace("-", "")
```

```python
num2 = df["ORDER_AMOUNT"].str.contains(",", regex=False).sum()
print("Number of records with ',' in the 'ORDER_AMOUNT' field:", num2)
```

```python
df["ORDER_AMOUNT"] = df["ORDER_AMOUNT"].replace(",", ".")
```

```python
num3= (df["ORDER_CREATION_DATE"].dt.date == df["REQUESTED_DELIVERY_DATE"].dt.date).sum()
print("Number of records where order date and delivery date are the same:", num3)
```

```python
currency_counts = df["ORDER_CURRENCY"].value_counts()
print(currency_counts)
```

```python
df
```

```python
df['ORDER_AMOUNT'] = pd.to_numeric(df['ORDER_AMOUNT'], errors='coerce')

usd_exchange_rate = {
    'USD': 1.0,   
    'EUR': 1.19,  
    'AUD': 0.75,   
    'CAD': 0.81,   
    'GBP': 1.39,  
    'MYR': 0.24,  
    'PLN': 0.27,   
    'AED': 0.27, 
    'HKD': 0.13,  
    'CHF': 1.09,  
    'RON': 0.24,   
    'SGD': 0.75, 
    'CZK': 0.045,
    'HU1': 0.0036, 
    'NZD': 0.71,   
    'BHD': 2.65,   
    'SAR': 0.27,   
    'QAR': 0.27,   
    'KWD': 3.33,   
    'SEK': 0.12
}

df['amount_in_usd'] = df.apply(lambda row: row['ORDER_AMOUNT'] / usd_exchange_rate[row['ORDER_CURRENCY']] if row['ORDER_CURRENCY'] != 'USD' else row['ORDER_AMOUNT'], axis=1)
print("Updated dataframe:")
print(df)
```
